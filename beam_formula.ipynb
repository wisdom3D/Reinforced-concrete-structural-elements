```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

def beam1(P,L,EI):
  R1 = P/2
  R2 = P/2
  Vmax= P/2
  Mmax = P*L/4
  deltamax = (P*L**3)/48*EI
  plt.style.use('seaborn-whitegrid')
  X = np.linspace(0,L/2,10)
  # for x < L/2
  Mx = P*X/2
  # for  L/2 <= x <= L   
  X1 = np.linspace(L/2,L,10)
  MX1 = 0.5*P*X1 - P*(X1 - L/2)
#--------------------------------------------------
  fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(23,5))
  #plt.plot([0]*L, color='c')
  ax1.plot(X, Mx, color='c')
  ax1.plot(X1, MX1, color='c')
  ax1.fill_between(X, Mx, alpha = 0.4, color='r')
  ax1.fill_between(X1, MX1, alpha = 0.4, color='r')
  ax1.set_title("MOMENTS DE FLEXION")
  ax1.plot([0, L], [0,0], linewidth = 2, color = "black", marker = 6, markersize = 16)
#------------------------------------------------------
  ax2.plot([0,L/2], [Vmax,Vmax], color='c')
  ax2.plot([L/2,L], [-Vmax,-Vmax], color='c')
  ax2.fill_between([0,L/2], [Vmax,Vmax], alpha = 0.4, color='r')
  ax2.fill_between([L/2,L], [-Vmax,-Vmax], alpha = 0.4, color='r')
  ax2.set_title("EFFORT TRANCHANT")
  ax2.plot([0, L], [0,0], linewidth = 2, color = "black", marker = 6, markersize = 16)
  return f'R1 = {R1} , R2 = {R2}, Vmax = {Vmax}, Mmax = {Mmax}, deltamax = {deltamax}'
# put the values of P, L and EI in the following function
#example  beam1(P,L,EI)
beam1(20, 4, 1200)
```

```python
def beam2(P,a,b,EI):
  R1 = P*b/(a+b)
  R2 = P*b/(a+b)
  Vmax= max(R1, R2)
  Mmax = P*a*b/(a+b)
  deltamax = (P*a*b*(a+2*b)*math.sqrt(3*a*(a+2*b)))/27*EI*(a+b)
  f = P*(a+b)**3/48*EI
  X = np.linspace(0,a,10)
  # for x < a
  Mx = (P*b*X)/(a+b)
  # for  x > a 
  X1 = np.linspace(a,(a+b),10)
  MX1 = (P*b*X1)/(a+b) - P*(X1 - a)
#--------------------------------------------------
  fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(23,5))
  plt.style.use('seaborn-whitegrid')
  #plt.plot([0]*(a+b), color='c')
  ax1.plot(X, Mx, color='c')
  ax1.plot(X1, MX1, color='c')
  ax1.fill_between(X, Mx, alpha = 0.4, color='r')
  ax1.fill_between(X1, MX1, alpha = 0.4, color='r')
  ax1.set_title("MOMENTS DE FLEXION")
  ax1.plot([0, (a+b)], [0,0], linewidth = 2, color = "black", marker = 6, markersize = 16)
#------------------------------------------------------
  ax2.plot([0,a], [Vmax,Vmax], color='c')
  ax2.plot([a,(a+b)], [-Vmax,-Vmax], color='c')
  ax2.fill_between([0,a], [Vmax,Vmax], alpha = 0.4, color='r')
  ax2.fill_between([a,(a+b)], [-Vmax,-Vmax], alpha = 0.4, color='r')
  ax2.set_title("EFFORT TRANCHANT")
  ax2.plot([0, (a+b)], [0,0], linewidth = 2, color = "black", marker = 6, markersize = 16)
  return f'R1 = {R1} , R2 = {R2}, Vmax = {Vmax}, Mmax = {Mmax}, deltamax = {deltamax}, flche = {f}'
# put the values of P, a, b, and EI in the following function
#example  beam2(P,a,b,EI)
beam2(10, 8, 3, 1200)
```

```python
def beam3(W,a,b,c,EI):
  R1 = W*b*(2*c+b)/(2*(a+b+c))
  R2 = W*b*(2*a+b)/(2*(a+b+c))
  Vmax = max(R1, R2)
  Mmax = R1*(a+R1/2*W)
  deltamax = (5*W*(a+b+c)**4)/384*EI
 
  X0 = np.linspace(a,(a+b),10)
  Vx = R1-W*(X0-a)

  X = np.linspace(0,a,8)
  X1 = np.linspace(a,(a+b),8)
  X2 = np.linspace((a+b),(a+b+c),8)
# for x < a
  Mx = R1*X
# for a < x < a+b
  Mx1 = R1*X1 - (W*(X1-a)**2)/2
# for x >a+b
  Mx2 = R2*((a+b+c)-X2)
  fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(23,5))
  plt.style.use('seaborn-whitegrid')
  ax1.plot(X, Mx, color='c')
  ax1.fill_between(X, Mx, alpha = 0.4, color='r')
  ax1.plot(X1, Mx1, color='c')
  ax1.fill_between(X1, Mx1, alpha = 0.4, color='r')
  ax1.plot(X2, Mx2, color='c')
  ax1.fill_between(X2, Mx2, alpha = 0.4, color='r')
  ax1.set_title("MOMENTS DE FLEXION")
  ax1.plot([0, (a+b+c)], [0,0], linewidth = 2, color = "black", marker = 6, markersize = 16)
#-----------------------------------------------------------
  ax2.plot([0,a], [Vmax,Vmax], color='c')
  ax2.fill_between([0,a], [Vmax,Vmax], alpha = 0.4, color='r')
  ax2.plot(X0, Vx, color='c')
  ax2.fill_between(X0, Vx, alpha = 0.4, color='r')
  ax2.plot([(a+b),(a+b+c)], [-Vmax,-Vmax], color='c')
  ax2.fill_between([(a+b),(a+b+c)],[-Vmax,-Vmax], alpha = 0.4, color='r')
  ax2.set_title("EFFORT TRANCHANT")
  ax2.plot([0, (a+b+c)], [0,0], linewidth = 2, color = "black", marker = 6, markersize = 16)
  return f'R1 = {R1} , R2 = {R2}, Vmax = {Vmax}, Mmax = {Mmax}, deltamax = {deltamax}'
# put the values of W, a, b, c,and EI in the following function
#example  beam2(W,a,b,c,EI)
beam3(12, 8, 5, 12, 1200)
```
